```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Womens Clothing E-Commerce Reviews.csv')
```

```python
df.drop(df.columns[0],inplace=True, axis=1)
```

```python
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.stem.lancaster import LancasterStemmer
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

```python
tokenizer = RegexpTokenizer(r'\w+')
```

```python
stop_words = set(stopwords.words('english'))

def preprocessing(data):
    txt = data.str.lower().str.cat(sep=' ') #1
    words = tokenizer.tokenize(txt) #2
    words = [w for w in words if not w in stop_words] #3
    #words = [ps.stem(w) for w in words] #4
    return words
```

```python
from nltk.stem.porter import PorterStemmer
#ps = LancasterStemmer()
ps = PorterStemmer()
```

```python
df['tokenized'] = df["Review Text"].astype(str).str.lower() # Turn into lower case text
df['tokenized'] = df.apply(lambda row: tokenizer.tokenize(row['tokenized']), axis=1) # Apply tokenize to each row
df['tokenized'] = df['tokenized'].apply(lambda x: [w for w in x if not w in stop_words]) # Remove stopwords from each row
df['tokenized'] = df['tokenized'].apply(lambda x: [ps.stem(w) for w in x]) # Apply stemming to each row
```

```python
df.head()
```

```python
from gensim.models import Word2Vec
```

```python
# train model
model = Word2Vec(df['tokenized'], min_count=1)
# summarize the loaded model
print(model)
# summarize vocabulary
words = list(model.wv.vocab)
print(words)
```

```python
from sklearn.manifold import TSNE
```

```python
viz_words = 500
tsne = TSNE()
X = model[model.wv.vocab]
transformed = tsne.fit_transform(X[:viz_words, :])
```

```python
fig, ax = plt.subplots(figsize=(16, 16))

plt.scatter(transformed[:, 0], transformed[:, 1])
for i, word in enumerate(words[:viz_words]):
    plt.annotate(word, xy=(transformed[i, 0], transformed[i, 1]), alpha=0.7)
plt.show()
```
